```python
fig=sns.heatmap(ablone.corr(), annot=True, cmap=sns.cubehelix_palette(as_cmap=True))
fig.set(xlabel='DIMENSIONS', ylabel='Dims', title='CORRELATION MATRIX')
```

```python
fig1=sns.scatterplot(data=ablone, x='Length',y='Diameter', alpha=0.7,hue='Sex',palette='Set1')
fig1.set(title='LENGTH VS DIAMETER')
```

```python
fig1=sns.scatterplot(data=ablone[ablone['Sex']=='F'], x='Length',y='Diameter')
fig1.set(title='LENGTH VS DIAMETER FOR FEMALE ABLONES')
```

```python
fig2=sns.boxplot(y='Viscera weight', x='Sex', data=ablone)
fig2.set(title='BOXPLOT')
```

```python
fig1=sns.scatterplot(data=ablone, x='Shucked weight',y='Rings', alpha=0.7)
fig1.set(title='LEEAST CORRELATED VARIABLES')
```
