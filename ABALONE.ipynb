```python
#obtain correleation matrix 
abalone_without_sex_Col=abalone[['Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight',
       'Viscera weight', 'Shell weight', 'Rings']]
abalone_without_sex_Col.corr()
```

```python
fig=sns.heatmap(abalone_without_sex_Col.corr(), annot=True, cmap=sns.cubehelix_palette(as_cmap=True))
fig.set(xlabel='DIMENSIONS', ylabel='Dims', title='CORRELATION MATRIX')
```

```python
fig1=sns.scatterplot(data=abalone, x='Length',y='Diameter', alpha=0.7,hue='Sex',palette='Set1')
fig1.set(title='LENGTH VS DIAMETER')
```

```python
fig2=sns.scatterplot(data=abalone[abalone['Sex']=='F'], x='Length',y='Diameter')
fig2.set(title='LENGTH VS DIAMETER FOR FEMALE ABLONES')
```

```python
fig3=sns.boxplot(y='Viscera weight', x='Sex', hue='Sex', data=abalone)
fig3.set(title='DISTRIBUTION OF VISCERA WEIGHT FOR 3 DIFFERENT SEXES')
```

```python
#visualising the least correlated variables based on correlation matrix
fig4=sns.scatterplot(data=abalone, x='Shucked weight',y='Rings', alpha=0.7)
fig4.set(title='LEAST CORRELATED VARIABLES')
```

### Summary of EDA


1. Dataset is complete without any null/ empty values. This can be readily used for EDA.
2. There is only one non-numeric column ('Sex').
3. Standard deviation is high for 'Rings' column (3.224169) compared to other columns. This can suggest possibility of more outliers in 'Rings' column.
4. 'Length' and 'Diameter' are most correleated variables in the dataset (Correleation coefficient = 0.99). This infers that both the variables are highly related.
5. 'Shucked weight' and 'Rings' are least correlated variables (Correleation coefficient = 0.42). This infers that there is no good relation between these 2 variables. Considering them together to derive insights would not be of use.
